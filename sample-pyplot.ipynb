```python
def count(data: list) -> int:
    count = 0
    for d in data:
        if (d % 3 == 0) and (d % 5) != 0:
            count += 1
    return count
```

```python
data = [15, 4, 27, 29, 9]
print(count(data))
```

```python
x = [48, 240, 557, 185, 655, 475, 250, 177, 132, 598, 979, 479, 748, 672, 558]
print(count(x))
```
