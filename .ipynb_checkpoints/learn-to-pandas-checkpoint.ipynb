## Merge

```python
# 名前のデータフレーム定義
df_name = pd.DataFrame(
    {
        "key":[1, 1, 3, 2, 5],
        "name":["asuka", "yuuki", "siho", "rina", "manaka"]
     }
     )
df_name
```

```python
df_group = pd.DataFrame(
    {
        "key":[1, 2, 3, 4],
        "group":["nogi", "sakura", "hinata", "yosimoto"]
     }
     )
df_group
```

```python
# 内部結合
pd.merge(df_name, df_group, how="inner")
```

```python
# 内部結合
pd.merge(df_name, df_group, how="left")
```

```python
pd.merge(df_name, df_group, how="right")
```

```python
pd.merge(df_name, df_group, how="outer")
```

```python
df_name_1 = pd.DataFrame(
    {
        "key":[1, 2],
        "name":["asuka", "rina"]
    }
     )
df_name_1
```

```python
df_name_2 = pd.DataFrame(
    {
        "key":[1, 2, 3],
        "name":["hinako", "yui", "kyouko"]
     }
     )
df_name_2
```

```python
df_group_1 = pd.DataFrame(
    {
         "group": ["nogi", "sakura", "nogi", "sakura", "hinata"]
    }
)
df_group_1
```

```python
# axis=0 => 縦結合
df_name_3 = pd.concat(
    [df_name_1, df_name_2], 
    axis=0,
    ignore_index=True
)
df_name_3
```

```python
# あんま良くないやり方(ignore_indexなし)
df_name_4 = pd.concat(
    [df_name_1, df_name_2], 
    axis=0
)
df_name_4
```

```python
# Reset!
df_name_4.reset_index(drop=True)
```

```python
# axis=1 => 横結合
df_name_group_2 = pd.concat([df_name_3, df_group_1], axis=1)
df_name_group_2
```

## 集約メソッド

```python
# クラス，性別，身長のデータフレーム
df = pd.DataFrame(
  {
    "class":["a", "b", "c", "c", "a", "c", "b", "a", "c", "b", "a"],
    "gender":["M", "F", "F", "M", "F", "M", "M", "F", "M", "M", "F"],
    "height":[162, 150, 168, 173, 162, 198, 182, 154, 175, 160, 172]
  }
  )
```

```python
df
```

```python
# classでグループイングして、heightの平均を算出
df.groupby("class").mean()
```

```python
df.groupby("class", as_index=False).mean()
```

```python
df.groupby("class", as_index=False).std()
```

```python
df.groupby(["class", "gender"], as_index=False).mean()
```

```python
# グルーピングしたデータフレームを取得
df.groupby("class").get_group("a")
```

```python
# 基本統計量の算出
df.groupby("class").describe()
```

## 縦変換 / 横変換

```python
# wide型のデータフレームを定義
df_wide = pd.DataFrame(
  {
    "store":["A", "B", "C"],
    "orange":[100, 70, 120],
    "apple":[150, 90, 80]
  }
)
df_wide
```

```python
# wide型をlong型に変換
df_long = df_wide.melt(
    id_vars="store",
    value_vars=["orange", "apple"],
    var_name="furit",
    value_name="price"
)
df_long
```

```python
# long型をwide型に変換
df_long.pivot(
    index="store",
    columns="furit",
    values="price"
).reset_index()
```

## 繰り返し処理

```python
series = sns.load_dataset('diamonds')['color']
series
```

```python
# 関数
def func(x):
    if x == "E":
        return("E")
    elif x == "I":
        return("I")
    else:
        return("not_E_I")
```

```python
func("E")
```

```python
# mapに関数を指定すると繰り返し処理をしてくれる
series.map(func)
```

```python
df = \
    sns \
    .load_dataset('diamonds') \
    .filter(["depth", "table", "color"])

df
```

```python
# assignはフレームに列を追加する
df_rev = df.assign(color_rev = lambda df : df["color"].map(func) )
df_rev
```

```python
# 辞書使用
# E => e
# I = i
# not_E_I => not_e_i
df_rev["color_rev"].map(
    {
        "E": "e",
        "I": "i",
        "not_E_I":"not_e_i"
    }
)
```

## 欠損値処理

```python
# Create DataFrame
df = pd.DataFrame(
{
    "x": [1, 3, 2, 8, 10],
    "y": [10, 4, np.nan, 6, np.nan],
    "z": [np.nan, 8, 7, 8, np.nan]
})
df
```

```python
df.dropna()
```

```python
# z列に欠損値がある行を削除
df.dropna(subset=["z"])
```

```python
df = pd.DataFrame(
{
    "x": [1, 3, 2, 8, 10],
    "y": [10, 4, np.nan, 6, np.nan],
    "z": [np.nan, 8, 7, 8, np.nan]
}).astype("float")
df
```

```python
## value: 置換後の値を直接指定
# 欠損値を0にする
df.fillna(value=0)
```

```python
# valueを省略できる
df.fillna(0)
```

```python
# y列の欠損値を5に、z列の欠損値を3に変換
df.fillna({"y": 5, "z": 3})
```

```python
# 欠損値を各列の平均に置換
df.fillna(df.mean())
```

```python
# 各列の平均
df.mean()
```

```python
## method: 置換後の値を欠損値の上下の値に指定

# 欠損値を上の値で置換
df.fillna(method="ffill")
```

```python
# 欠損値を下の値で置換
df.fillna(method="bfill")
```
