```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('data/income_and_expenditure.csv')
df.head(3)
```

## Q1. 世帯収入（変数名：Total Household Income）に対して、相関係数の絶対値が 0.5 を超える特徴量をすべて選択せよ

```python
# 相関係数の確認
df_corr = df.corr()
df_corr
```

```python
plt.figure(figsize=(12, 8))
sns.heatmap(df_corr.iloc[:20, :20], annot=True, cmap='Blues')
```

```python
# 絶対値の取得
df_abs = df_corr.abs()

# Total Household Income列の取得
df_Total_Household_Income = df_abs['Total Household Income']

df_Total_Household_Income[df_Total_Household_Income > 0.5]
# Transportation Expenditure     
# Communication Expenditure
# Restaurant and hotels Expenditure
```
