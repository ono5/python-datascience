```python
import pandas as pd
import numpy as np
import seaborn as sns
from plotnine import *
from plotnine.themes import *
import warnings
 
# データフレームの表示行数を指定
pd.set_option('display.max_rows', 5) 
 
# 図のサイズ調整
theme_set(theme(
    figure_size=(4.2, 3.6), 
    axis_title=element_text(size=15), 
    axis_text=element_text(size=15)
    )
)
 
# 警告の非表示
warnings.filterwarnings('ignore')
```

```python
df = \
sns\
.load_dataset('iris')
df
```

```python
# 1.キャンパス
ggplot(
  data = df, 
  mapping = aes(x="sepal_length", y="petal_length")
)
```

```python
# 1 キャンパス + 2. グラフ
ggplot(
  data = df, 
  mapping = aes(x="sepal_length", y="petal_length")
) + \
geom_point() # 点
```

```python
# 1 キャンパス + 2. グラフ + 3. 体裁
# 1 キャンパス + 2. グラフ
ggplot(
  data = df, 
  mapping = aes(x="sepal_length", y="petal_length")
) + \
geom_point() + \
theme(
    rect=element_rect(fill="pink", color="orange", size=5),
    axis_title=element_text(size=30),
    axis_text=element_text(size=30)
)
```

```python
# カテゴリ(単数)
ggplot(
    df, 
    aes(x="sepal_length", fill="species")
) + \
geom_histogram(position="identity", alpha=0.6)
```

```python
# sepal_lengthのヒストグラム カテゴリごと出力 (複数)
ggplot(
    df,
    aes(x="sepal_length")
) + \
geom_histogram() + \
facet_wrap("species")
```

## geom_bar

```python
# speciesでグルーピングして各列の平均値を算出
df_group = df.groupby("species").mean().reset_index()
df_group
```

```python
# 各カテゴリに対するspeal_lengthの平均値の棒グラフ
ggplot(
    df_group,
    aes(x="species", y="sepal_length")
) + \
geom_bar(stat="identity")
```

```python
# 各カテゴリに対するspeal_lengthの平均値の棒グラフ(fillを使用)
ggplot(
    df_group,
    aes(x="species", y="sepal_length", fill="species")
) + \
geom_bar(stat="identity")
```

```python
# 各カテゴリに対するspeal_lengthの平均値の棒グラフ(縦軸と横軸を入れ替える)
ggplot(
    df_group,
    aes(x="species", y="sepal_length")
) + \
geom_bar(stat="identity") + \
coord_flip()
```

```python
# sepal_length以外の平均値も比較する
# df_groupをlong型に変換
df_group_long = df_group.melt(
    id_vars="species",
    value_vars=["sepal_length", "sepal_width", "petal_length", "petal_width"]
)
df_group_long
```

```python
# 各カテゴリに対する前列の平均値の棒グラフ
ggplot(
    df_group_long,
    aes(x="species", y="value", fill="species")
) + \
geom_bar(stat="identity") + \
coord_flip() + \
facet_wrap("variable")
```

## geom_point(散布図)

```python
# 散布図 x:sepal_length y: sepal_width
ggplot(
    df,
    aes(x="sepal_length", y="sepal_width")
) + \
geom_point()
```

```python
# 散布図 x:sepal_length y: sepal_width カテゴリごと(単数)
ggplot(
    df,
    aes(x="sepal_length", y="sepal_width", color="species")
) + \
geom_point()
```

```python
# 散布図 x:sepal_length y: sepal_width カテゴリごと(複数)
ggplot(
    df,
    aes(x="sepal_length", y="sepal_width", color="species")
) + \
geom_point() + \
facet_wrap("species")
```
